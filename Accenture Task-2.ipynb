# Loading the Dataset

```python
import numpy as np
import pandas as pd
```

```python
content = pd.read_csv('Content.csv')
reaction = pd.read_csv('Reactions.csv')
reactiontypes = pd.read_csv('ReactionTypes.csv')
```

```python
content.head()
```

```python
reaction.head()
```

```python
reactiontypes.head()
```

# Data Cleaning and Modelling

```python
content.isnull().sum()
```

```python
content.drop(['User ID'],axis=1,inplace=True)
```

```python
content
```

```python
content['Category'].unique()
```

```python
content['Category'] = content['Category'].str.replace('"', '', regex=False)
content['Category'] = content['Category'].str.lower().str.strip()
content['Category'].unique()
```

```python
content
```

```python
reaction.isnull().sum()
```

```python
reaction.drop(['Unnamed: 0','User ID'],axis=1,inplace=True)
```

```python
reaction.isnull().sum()
```

```python
reaction.dropna(inplace=True)
```

```python
reaction
```

```python
reaction.isnull().sum()
```

```python
merged_df = pd.merge(reaction, content, how='left', on='Content ID')
merged_df
```

```python
merged_df.rename(columns={"Type_x":"Reaction Type","Type_y":"Content Type"},inplace=True)
```

```python
merged_df
```

```python
reactiontypes
```

```python
reactiontypes.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
reactiontypes.rename(columns={"Type":"Reaction Type"},inplace=True)
```

```python
reactiontypes
```

```python
final = pd.merge(merged_df,reactiontypes,on="Reaction Type",how="left")
```

```python
final
```

```python
final['Category'].unique()
```

# Top Performing Categories

```python
top_categories = final.groupby(['Category'])['Score'].sum().reset_index()
top_categories
```

# Saving Our Work 

```python
final.to_csv('final.csv')
top_categories.to_csv('top_categories.csv')
```
